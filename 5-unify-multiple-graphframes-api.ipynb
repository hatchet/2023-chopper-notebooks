```python
import hatchet as ht

g_toy1 = [
    {
        "frame": {"name": "foo", "type": "function"},
        "metrics": {"time (inc)": 16.0, "time": 6.0},
        "children": [
            {
                "frame": {"name": "bar"},
                "metrics": {"time (inc)": 9.0, "time": 4.0},
                "children": [
                    {
                        "frame": {"name": "baz", "type": "function"},
                        "metrics": {"time (inc)": 3.0, "time": 3.0},
                    },
                ],
            },
            {
                "frame": {"name": "baz"},
                "metrics": {"time (inc)": 1.0, "time": 1.0},
            },
        ],
    },
]

g_toy2 = [
    {
        "frame": {"name": "foo", "type": "function"},
        "metrics": {"time (inc)": 8.0, "time": 2.0},
        "children": [
            {
                "frame": {"name": "baz"},
                "metrics": {"time (inc)": 6.0, "time": 6.0},
            },
        ],
    },
]

g_toy3 = [
    {
        "frame": {"name": "foo", "type": "function"},
        "metrics": {"time (inc)": 16.0, "time": 0.0},
        "children": [
            {
                "frame": {"name": "bar"},
                "metrics": {"time (inc)": 16.0, "time": 5.0},
                "children": [
                    {
                        "frame": {"name": "baz", "type": "function"},
                        "metrics": {"time (inc)": 11.0, "time": 8.0},
                        "children" : [
                            {
                                "frame" : {"name": "garply", "type": "function"},
                                "metrics": {"time (inc)": 3.0, "time": 3.0},
                            }
                        ]
                    },
                ],
            },
        ],
    },
]

gf_toy1 = ht.GraphFrame.from_literal(g_toy1)
gf_toy1.update_metadata(1)
gf_toy2 = ht.GraphFrame.from_literal(g_toy2)
gf_toy2.update_metadata(2)
gf_toy3 = ht.GraphFrame.from_literal(g_toy3)
gf_toy3.update_metadata(3)
gf_toys = [gf_toy1, gf_toy2, gf_toy3]

ht.GraphFrame.unify_multiple_graphframes(gf_toys)
print(gf_toy1.tree(), gf_toy2.tree(), gf_toy3.tree())
```
