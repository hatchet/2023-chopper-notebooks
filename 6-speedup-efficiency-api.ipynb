```python
import hatchet as ht
import glob
import re

profiles = glob.glob(
    "/home/ocankur/research/research-code/hatchet-aditya-tomar/hatchet-tutorial/data/lulesh-scaling/chopper-example/*"
)

gfs_eff_simple = ht.GraphFrame.construct_from(profiles)
gfs_eff_simple_flt = []

profile_gf = dict(zip(profiles, gfs_eff_simple))
for profile, gf in profile_gf.items():
    num_pes = re.match("(.*)-(\d+)(.*)", profile).group(2)
    gf.update_metadata(int(num_pes))
    gfs_eff_simple_flt.append(gf.filter(lambda x: not x['name'].startswith('MPI')))
    
lulesh_simple_weak = ht.Chopper.speedup_efficiency(gfs_eff_simple_flt, metrics=["time"], weak=True, efficiency=True)
display(lulesh_simple_weak.head(3))
```
