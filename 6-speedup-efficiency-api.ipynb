```python
import hatchet as ht
import glob

dirs_simple_lulesh = glob.glob("/home/ocankur/research/research-code/hatchet-tutorial/data/lulesh-scaling/*", recursive=True) 

pes = ["64", "125", "216", "512"]
pes_order = []
data = []
for d in dirs_simple_lulesh:
    pe = d.split("-")[-1].split("c")[0]
    if pe in pes:
        data.append(d)
        pes_order.append(int(pe))

gfs_eff_simple = ht.GraphFrame.construct_from(data)
        
for idx in range(len(gfs_eff_simple)):
    gf = gfs_eff_simple[idx]
    gf.update_metadata(pes_order[idx])
    gfs_eff_simple[idx] = gf.filter(lambda x: not x['name'].startswith('MPI'))
    
lulesh_simple_weak = ht.Chopper.speedup_efficiency(gfs_eff_simple, metrics=["time"], weak=True, efficiency=True)
display(lulesh_simple_weak.head(3))
```
