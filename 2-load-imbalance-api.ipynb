```python
import hatchet as ht

gf_imb_simple = ht.GraphFrame.from_caliper("/home/ocankur/research/research-code/hatchet-tutorial/data/lulesh-16nodes/lulesh-annotation-profile-512cores-nompi.json")
gf_imb_simple = gf_imb_simple.load_imbalance(verbose=True)

# The rest is styling for the paper.
gf_imb_simple.dataframe.reset_index(inplace=True)
gf_imb_simple.dataframe.set_index("name", inplace=True)
gf_imb_simple.dataframe['time.mean'] = gf_imb_simple.dataframe['time.mean'].apply(lambda x: '%.2E' % x)
gf_imb_simple.dataframe['time.percentiles'] = gf_imb_simple.dataframe['time.percentiles'].apply(lambda x: ['%.2E' % i for i in x])
gf_imb_simple.dataframe=gf_imb_simple.dataframe[["time.mean", "time.ranks", "time.hist", "time.percentiles", "time.imbalance"]]
gf_imb_simple.dataframe.head(3)
```
